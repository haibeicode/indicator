```python
from indicator.base import *
import tushare as ts
```

```python
data = ts.get_hist_data('600848')
CLOSE = data['close']
CLOSE.head()
```

```python
"""
通达信代码
DIF:EMA(CLOSE,SHORT)-EMA(CLOSE,LONG);
DEA:EMA(DIF,MID);
MACD:(DIF-DEA)*2,COLORSTICK;
"""
SHORT=12
LONG=2
MID=9
DIF=EMA(CLOSE,SHORT)-EMA(CLOSE,LONG)
DEA=EMA(DIF,MID)
MACD=(DIF-DEA)*2
```

```python
import pandas as pd

madc = pd.DataFrame()

madc['DIF'] = DIF
madc['DEA'] = DEA
madc['MACD'] = MACD
# MADC = {"DIF":DIF,"DEA":DEA,"MACD":MACD}
madc
```
